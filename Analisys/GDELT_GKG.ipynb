```python
import pandas as pd
from datetime import datetime
```

```python
df=pd.read_csv('20240811.gkg.csv.zip',compression='zip',header=0,sep='\t')
```

```python
df.info()
```

```python
def normalize_date(date):
    # Convert to datetime object
    date_obj = datetime.strptime(str(date), "%Y%m%d")

    # Convert to desired string format
    normalize_date = date_obj.strftime("%Y-%m-%d")

    return normalize_date
```

```python
# Now we apply this function to de dataframe[DATE] column
df['DATE'] = df['DATE'].apply(normalize_date)
```

```python
# We separate explode the TONE column so we can have it in diferent columns
df['TONE'] = df['TONE'].apply(lambda x: x.split(','))

# We create a new column for each value in the TONE column
df['AVG_TONE'] = df['TONE'].apply(lambda x: x[0])
df['POS_SCORE'] = df['TONE'].apply(lambda x: x[1])
df['NEG_SCORE'] = df['TONE'].apply(lambda x: x[2])
df['POLARITY'] = df['TONE'].apply(lambda x: x[3])
df['ARD'] = df['TONE'].apply(lambda x: x[4])
df['SGRD'] = df['TONE'].apply(lambda x: x[5])

# Drop the TONE column
df.drop(columns=['TONE'], inplace=True)
```

```python
# We turn the THEMES column into a list of strings
df['THEMES'] = df['THEMES'].apply(lambda x: x.split(';') if isinstance(x, str) else [])
```

```python
# We do the same with the PERSONS column
df['PERSONS'] = df['PERSONS'].apply(lambda x: x.split(';') if isinstance(x, str) else [])
```

```python
# And for the ORGANIZATIONS column too
df['ORGANIZATIONS'] = df['ORGANIZATIONS'].apply(lambda x: x.split(';') if isinstance(x, str) else [])
```
